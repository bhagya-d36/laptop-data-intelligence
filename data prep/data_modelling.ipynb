```python
# Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Importing the laptop store information
df_store = pd.read_csv('laptop_store_info.csv')
df_store
```

```python
# Importing the laptop technical specifications of the 4 laptops
df_specs = pd.read_csv('laptop_tech_specs.csv')
df_specs
```

```python
# Standardize the 'Model' column in both DataFrames to ensure consistency
df_specs.loc[df_specs['Brand'] == 'HP', 'Model'] = df_specs.loc[df_specs['Brand'] == 'HP', 'Model'].str.replace('^HP ', '', regex=True)
df_specs['Model'].value_counts()
```

```python
# Merging the two DataFrames on the 'Brand' and 'Model' columns
df = pd.merge(df_store, df_specs, on=['Brand', 'Model'], how='outer')
df
```

```python
# Check for missing values in the combined DataFrame
pd.isnull(df).sum()
```

```python
# Fill missing values with a placeholder ('-')
df = df.fillna('-')
pd.isnull(df).sum()
```

```python
# Save the cleaned and combined DataFrame to a new CSV file
df.to_csv(r'C:\Users\Bhagya Dissanayake\Desktop\ai-engineer-assessment-new\data\processed\laptop_info.csv', index=False)
```
