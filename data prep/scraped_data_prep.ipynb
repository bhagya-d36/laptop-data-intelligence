```python
import numpy as np
import pandas as pd
from pprint import pprint
```

```python
df_main = pd.read_csv('AI_scraped_data_4.csv')
df_main.head()
```

```python
# To better view the nested details
pprint(df_main.to_dict('records')[0])
```

```python
df_hp = pd.read_csv('AI_scraped_data_hp.csv')
df_hp
```

```python
df_len = pd.read_csv('AI_scraped_data_len.csv')
df_len
```

##### Data Standardization

```python
#df_main.loc[df_main['Brand'] == 'HP', 'Model'] = 'HP ' + df_main.loc[df_main['Brand'] == 'HP', 'Model']
df_main.head()
```

```python
#df_hp['Model'] = 'HP ' + df_hp['Model']
df_hp['Availability'] = 'Available'
df_hp.head()
```

```python
# Add "Available" to all entries in the availability column for df_len
df_len['Availability'] = 'Available'
df_len.head()
```

```python
# Check if all three DataFrames have the same columns
print("df_main columns:", set(df_main.columns))
print("df_hp columns:", set(df_hp.columns))
print("df_len columns:", set(df_len.columns))

if set(df_main.columns) == set(df_hp.columns) == set(df_len.columns):
    print("All three DataFrames have the same columns.")
else:
    print("The DataFrames have different columns.")
```

```python
# Combine all three DataFrames
df = pd.concat([df_main, df_hp, df_len], ignore_index=True)

# Display the combined DataFrame
display(df)
```

```python
# Check for missing values in the combined DataFrame
pd.isnull(df).sum()
```

```python
# Fill missing values with a placeholder ('-')
df = df.fillna('-')
pd.isnull(df).sum()
```

```python
# Save the cleaned and combined DataFrame to a new CSV file
df.to_csv('laptop_store_info.csv', index=False)
```
